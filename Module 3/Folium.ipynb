```python
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (math.sin(dlat/2))**2 + math.cos(lat1) * math.cos(lat2) * (math.sin(dlon/2))**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of earth in kilometers. Use 3956 for miles
    r = 6371 
    return c * r
```

```python
# Pick two launch sites
site1 = launch_sites.iloc[0]
site2 = launch_sites.iloc[1]

distance_km = haversine(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
print(f"Distance between {site1['Launch Site']} and {site2['Launch Site']}: {distance_km:.2f} km")
```

```python
distances = []
for i, site1 in launch_sites.iterrows():
    for j, site2 in launch_sites.iterrows():
        if i != j:
            dist = haversine(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
            distances.append({
                'Site1': site1['Launch Site'],
                'Site2': site2['Launch Site'],
                'Distance_km': dist
            })

distances_df = pd.DataFrame(distances)
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
lat1, lon1 = 28.5623, -80.5774   # Kennedy Space Center
lat2, lon2 = 34.6321, -120.6106  # Vandenberg Air Force Base

dist_km = calculate_distance(lat1, lon1, lat2, lon2)
print(f"Distance: {dist_km:.2f} km")
```

```python
import folium
from folium.features import DivIcon

# Coordinates (example)
launch_site_lat, launch_site_lon = 28.5623, -80.5774    # Launch site
coastline_lat, coastline_lon = 28.56367, -80.57163      # Closest coastline point

# Calculate distance function (make sure you have this defined already)
def calculate_distance(lat1, lon1, lat2, lon2):
    from math import sin, cos, sqrt, atan2, radians
    R = 6373.0  # Earth radius in km

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create map centered between points
map_center = [(launch_site_lat + coastline_lat) / 2, (launch_site_lon + coastline_lon) / 2]
site_map = folium.Map(location=map_center, zoom_start=13)

# Add marker for launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site',
    icon=folium.Icon(color='green', icon='rocket', prefix='fa')
).add_to(site_map)

# Add marker for coastline point
folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup='Closest Coastline',
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(site_map)

# Add distance label near coastline point
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color: #d35400;"><b>{distance_coastline:.2f} KM</b></div>',
    )
)
distance_marker.add_to(site_map)

# Draw PolyLine between launch site and coastline
coordinates = [
    [launch_site_lat, launch_site_lon],
    [coastline_lat, coastline_lon]
]
lines = folium.PolyLine(locations=coordinates, weight=3, color='blue')
site_map.add_child(lines)

# Display map
site_map
```

```python
import folium
from folium.features import DivIcon

# Launch site coordinate example
launch_site_lat, launch_site_lon = 28.5623, -80.5774

# Example POIs: cities, railways, highways, etc.
points_of_interest = [
    {'name': 'Closest City', 'lat': 28.5690, 'lon': -80.5850},
    {'name': 'Closest Railway', 'lat': 28.5700, 'lon': -80.5900},
    {'name': 'Closest Highway', 'lat': 28.5600, 'lon': -80.5750}
]

# Your previously defined distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    from math import sin, cos, sqrt, atan2, radians
    R = 6373.0

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

# Create map centered around launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add launch site marker
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site',
    icon=folium.Icon(color='green', icon='rocket', prefix='fa')
).add_to(site_map)

# Loop over each point of interest
for poi in points_of_interest:
    distance = calculate_distance(launch_site_lat, launch_site_lon, poi['lat'], poi['lon'])
    
    # Add POI marker
    folium.Marker(
        location=[poi['lat'], poi['lon']],
        popup=f"{poi['name']}<br>Distance: {distance:.2f} KM",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(site_map)
    
    # Add distance label near POI
    distance_marker = folium.Marker(
        location=[poi['lat'], poi['lon']],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color: #d35400;"><b>{distance:.2f} KM</b></div>',
        )
    )
    distance_marker.add_to(site_map)
    
    # Draw a line connecting launch site and POI
    line = folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [poi['lat'], poi['lon']]],
        weight=3,
        color='blue'
    )
    site_map.add_child(line)

# Show map
site_map
```
